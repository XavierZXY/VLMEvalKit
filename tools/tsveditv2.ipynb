# operation induction
## json to tsv

```python
import json
import csv
import os
import sys
import random
random.seed(2024)
import copy
query_json = os.path.pardir + "/datasets/operator_induction/query.json"
support_json = os.path.pardir + "/datasets/operator_induction/support.json"
query_tsv = os.getcwd() + "/datasets/operator_induction/query.tsv"
support_tsv = os.getcwd() + "/datasets/operator_induction/support.tsv"

# Step 2: Open and read the JSON file
with open(query_json, "r", encoding="utf-8") as json_file:
    query_data = json.load(json_file)

with open(support_json, "r", encoding="utf-8") as json_file:
    support_data = json.load(json_file)

# Step 3: Open a new TSV file for writing
with open(query_tsv, "w", newline="", encoding="utf-8") as tsv_file:
    writer = csv.writer(tsv_file, delimiter="\t")

    # Step 5: Write the headers to the TSV file
    headers = ["index", "question", "answer", "operator", "image"]
    writer.writerow(headers)

    # Step 6: Iterate over each item in the JSON data
    for idx, item in enumerate(query_data):
        # Step 7: Extract the required fields
        row = [
            item.get("id", ""),
            item.get("question", ""),
            item.get("answer", ""),
            item.get("operator", ""),
            item.get("image", ""),
        ]
        # Step 8: Write the extracted fields to the TSV file
        writer.writerow(row)

with open(support_tsv, "w", newline="", encoding="utf-8") as tsv_file:
    writer = csv.writer(tsv_file, delimiter="\t")

    # Step 5: Write the headers to the TSV file
    headers = ["index", "question", "answer", "image"]
    writer.writerow(headers)

    # Step 6: Iterate over each item in the JSON data
    for idx, item in enumerate(support_data):
        # Step 7: Extract the required fields
        row = [
            item.get("id", ""),
            item.get("question", ""),
            item.get("answer", ""),
            item.get("image", ""),
        ]
        # Step 8: Write the extracted fields to the TSV file
        writer.writerow(row)
```

## image to base64

```python
import csv
import base64
import os
import sys
from PIL import Image
import json

# 获取当前工作目录
current_dir = os.path.abspath(os.path.pardir)

# 将上级目录添加到sys.path中
sys.path.append(current_dir)
# 现在可以导入上级目录中的包
from vlmeval.smp.vlm import *

# 读取TSV文件并解析数据
input_tsv = os.getcwd() + "/datasets/operator_induction/support.tsv"
output_tsv = os.getcwd() + "/datasets/operator_induction/support.tsv"

data = []
count = 0
with open(input_tsv, "r", encoding="utf-8") as tsvfile:
    tsvreader = csv.reader(tsvfile, delimiter="\t")
    headers = next(tsvreader)
    data.append(headers)  # 保留headers
    for row in tsvreader:
        # image_path = os.path.join(os.path.dirname(input_tsv), row[headers.index("image")].strip("[]").strip("'"))
        # row[headers.index("image")] = encode_image_file_to_base64(image_path)
        datadir = os.path.join(os.path.pardir, "datasets")
        row[headers.index("image")] = encode_image_file_to_base64(
            os.path.join(datadir, row[headers.index("image")].strip("[]").strip("'"))
        )
        data.append(row)

# # 将修改后的数据写回到一个新的TSV文件中
with open(output_tsv, "w", encoding="utf-8", newline="") as tsvfile:
    tsvwriter = csv.writer(tsvfile, delimiter="\t")
    tsvwriter.writerows(data)

print("TSV file with base64 encoded images has been created.")
```
